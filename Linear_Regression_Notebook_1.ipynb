```python
kaggle_data_df_short['Total Proficiency'] = kaggle_data_df_short['Average ELA Proficiency'] + kaggle_data_df_short['Average Math Proficiency']
```

```python
"""
In our dataset, we had two separate variables that could best predict ****primary!!!!! school education outcomes.
To deal with this, we decided to take the average between our two columns, and use this as our target (dependent)
variable. After we created this averge metric, we decided to drop the two individual columns.
"""

kaggle_data_df_short['Total Proficiency Average'] = (kaggle_data_df_short['Average ELA Proficiency'] + kaggle_data_df_short['Average Math Proficiency'])/2
```

```python
# kaggle_data_df_short.drop('Total Proficiency',axis=1,inplace=True)
kaggle_data_df_short.drop(['Average ELA Proficiency','Average Math Proficiency'],axis=1,inplace=True)
```

```python
kaggle_data_df_short = kaggle_data_df_short.rename(columns={'Strong_Family-Community_Ties_Per':'Strong_Family_Community_Ties_Per',
                                                            'Percent_Black__Hispanic':'Percent_Black_Hispanic'})
```

```python
predictors_v1 = ['Economic_Need_Index', 'Percent_ELL', 'Percent_Asian', 'Percent_Black',
       'Percent_Hispanic', 'Percent_Black_Hispanic', 'Percent_White',
       'Student_Attendance_Rate', 'Percent_of_Students_Chronically_Absent',
       'Rigorous_Instruction_Per', 'Collaborative_Teachers_Per',
       'Supportive_Environment_Per', 'Strong_Family_Community_Ties_Per']

titles = ['ind_var', 'r_squared', 'intercept', 'slope', 'p-value', 'normality (JB)']
values_dict = []
for idx, val in enumerate(predictors_v1):
#     print("Regression Analysis and Diagnostics for formula: Total_Proficiency_Average~" + val)
#     print("-------------------------------------------------------------------------------------")
    f = "Total_Proficiency_Average~"+val
    model = smf.ols(formula=f, data=kaggle_data_df_short).fit()
    X_new = pd.DataFrame({val: [kaggle_data_df_short[val].min(), kaggle_data_df_short[val].max()]})
    preds = model.predict(X_new)
    l = [val, model.rsquared, model.params[0], model.params[1], model.pvalues[1], sms.jarque_bera(model.resid)[0]]
    values_dict.append(dict(zip(titles,l)))
```
